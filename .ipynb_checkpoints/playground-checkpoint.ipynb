```python
#path = '/mnt/h/data_IceTracker_2020m1/300234066532030/20200126T180844/src.tiff'
pdir = '/mnt/d/data_IceTracker_2020m1/174640'
import data_util
for i in range(len(os.listdir(pdir))):
    if ('.pickle' in os.listdir(pdir)[i]): continue
    #path = os.path.join(pdir, os.listdir(pdir)[i], 'tgt_pln.tiff')
    #ds_i = gdal.Open(path)
    #ds_i = ds_i.GetRasterBand(1).ReadAsArray()
    #plt.imshow((np.abs(ds_i)), cmap='grey'); plt.show()
    #print(len(ds_i.GetGCPs()))

    path = os.path.join(pdir, os.listdir(pdir)[i], 'src_pln.tiff')
    ds_i = gdal.Open(path)
    #ds_i = ds_i.GetRasterBand(2).ReadAsArray()[::8,::8]
    #plt.imshow(np.abs(ds_i)); plt.show()
    srs = osr.SpatialReference(ds_i.GetGCPProjection())
    f1 = np.array([[e.GCPLine,e.GCPPixel]+list(data_util.srs_llh2xyz(srs, e.GCPY, e.GCPX, e.GCPZ)) for e in ds_i.GetGCPs()])
    f1 = scipy.interpolate.RBFInterpolator(f1[:,:2], f1[:,2:])
    tmp = np.stack(np.meshgrid(np.arange(0,ds_i.RasterYSize,10),np.arange(0,ds_i.RasterXSize),indexing='ij'),axis=-1)
    ds_i = f1(tmp.reshape((-1,2))).reshape((tmp.shape[0],tmp.shape[1],3))
    ds_i = data_util.srs_xyz2llh(srs, ds_i[...,0], ds_i[...,1], ds_i[...,2])
    plt.imshow(ds_i[0]); plt.colorbar(shrink=0.1); plt.show()
    plt.imshow(ds_i[1]); plt.colorbar(shrink=0.1); plt.show()
    plt.imshow(ds_i[2]); plt.colorbar(shrink=0.1); plt.show()
    print('next')

#path2 = '/mnt/h/data_IceTracker_2020m1/300234066532030/20200126T180844/src_grd.tiff'
#path2 = '/mnt/d/data_IceTracker_2020m1/174637/20200101T051701/src_pln.tiff'
#plt.imshow()
#ds_o = gdal.Open(path2)
```
