```python
import ASF
import os, itertools
from burst2safe import local2safe

import pandas as pd

buoy_data = pd.read_csv('IABP_LEVEL1_2020.csv')

buoy_id = buoy_data['BuoyID'].iloc[0]
buoy_data = buoy_data[buoy_data['BuoyID']==buoy_id]
```

```python
buoy_data
```

```python

scroller = ASF.BuoyScroller(buoy_id=buoy_id, buoy_data=buoy_data)
scroller.search(delta = 86400 * 10, offset = 86400*0.5)
#scroller.download_results(cred_user='mjeiouws',cred_pass='MORYCORis343!')
```

```python
resprop_l = [el.properties for el in itertools.chain.from_iterable(scroller.results.values())]

SLC_dict = {}
for resprop in resprop_l:
    entry = SLC_dict.setdefault(resprop['groupID'], {}).setdefault(resprop['burst']['subswath'], {}).setdefault(resprop['polarization'],{}).setdefault(resprop['burst']['burstIndex'], {})
    entry['DATA'] = scroller.label + '_downloads/' + resprop['fileID'] + '.tiff'
    entry['METADATA'] = scroller.label + '_downloads/' + resprop['fileID'] + '.xml'
SLC_dicts = [{key:val} for key, val in SLC_dict.items()]
```
